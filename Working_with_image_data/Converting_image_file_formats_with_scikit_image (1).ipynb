# Convert RGBA to RGB format

```python
image_rgba = imread('/content/drive/MyDrive/Colab Notebooks/Deep Learning Notebooks/mid_alpha.png')
image_rgba.shape
```

To convert rgba to rgb we use the function `rgba2rgb`

```python
# conversion rgba to rgb
image_rgb = rgba2rgb(image_rgba)
image_rgb.shape
```

```python
# imshow result
imshow(image_rgb)
```

summary: although image was converted from rgba to rgb, the transparency did not change.

# Converting RGB to grayscale image

```python
# converting rgb to gray
image_gray = rgb2gray(image)
image_gray.shape
```

```python
# print image
imshow(image_gray)
```

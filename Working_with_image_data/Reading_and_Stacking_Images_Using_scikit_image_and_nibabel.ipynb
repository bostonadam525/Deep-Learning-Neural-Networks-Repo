### Printing different RGB image channels

```python
# plot image
imshow(image)
```

### To print on red channel
* replace other color matrices with value of 0
* keep values in red channel
* matrix multiplication using a "red_filter" by the original image matrix.

```python
# print only red channel
red_filter = [1,0,0]
imshow(image*red_filter)
```

### To display green channel....
* Same process as above, padd the non-green channels with 0.
* matrix multplication

```python
# print only green channel
green_filter = [0,1,0]
imshow(image*green_filter)
```

### Display blue channel only
* same process as above, pad non-blue channels with 0.
* matrix multiplication

```python
# create blue filter and matrix multiply
blue_filter = [0,0,1]
imshow(image*blue_filter)
```

```python
# individual channel using cmap
# cmap set to specific channel
imshow(image[:,:,0], cmap='Reds')
```

```python
# same experiment as above with green or blue
imshow(image[:,:,0], cmap='Blues')
```

### Save as hdr file
* Need to convert `nii` file to `hdr` or high dynamic range file.
* NII format is created by Neuroimaging Informatics Technology Initiative. It is commonly used to store magnetic resonance imaging (MRI) data.
* example: https://www.kaggle.com/code/datark1/what-are-dcm-and-nii-files-and-how-to-read-them
* affine is the mapping between array (voxel) coordinates and mm (world) coordinates (source: https://nipy.org/nibabel/reference/nibabel.affines.html_

```python
nib.save(img, '/content/drive/MyDrive/Colab Notebooks/Deep Learning Notebooks/brain_test.hdr')

#/content/drive/MyDrive/Colab Notebooks/Deep Learning Notebooks/someones_epi.nii
```

### glob can retrieve multiple images at once

```python
# reading multiple images
from glob import glob
image_name = glob("*.png")
```

```python
image_name
```

```python
# loop to append
image_ = []
for i in image_name:
  image = imread(i)
  imshow(image)
  plt.show()
  image_.append(image)

final_images = np.array(image_).reshape(2,)
```
